```python
import pandas as pd
from email_validator import validate_email, EmailNotValidError
```

```python

ind = 0

def check_for_valid(email):
    global ind

    ind += 1

    if ind % 50 == 0:
        print(f'{ind} - email отработано')
    try:
        emailinfo = validate_email(email, check_deliverability=True)
        email = emailinfo.normalized
        return 'valid and normalized'
    except EmailNotValidError as e:
        return f'is invalid: {str(e)}'  # Возвращаем сообщение об ошибке для отладки
    except Exception as e:
        return f'Error: {str(e)}' # Ловим другие ошибки



df = pd.read_excel('Продажи по контрагентам обработаный.xlsx', sheet_name='Sheet1')
df['check_for_valied_email'] = df['Email'].apply(check_for_valid)
df.to_excel('Продажи по контрагентам обработаный с валидностью емаилов.xlsx')
```
